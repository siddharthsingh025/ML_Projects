#**Popularity Based recommander system**

> *top 50 books having more than 250 vots and highest avg rating*

```python
mr_ratings=ratings.merge(books,on='ISBN')
mr_ratings
```

> *find rating number* : its actually counting vots

```python
nrating_df=mr_ratings.groupby('Book-Title').count()['Book-Rating'].reset_index()
nrating_df.rename(columns={'Book-Rating':'rating_num'},inplace=True)
nrating_df
```

> *Now we have to calculate avarage rating of books*

```python
avg_rating_df=mr_ratings.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_rating_df
```

```python
popularity_df = nrating_df.merge(avg_rating_df,on='Book-Title')
popularity_df
```

> *here we done this : top 50 books having more than 250 vots and highest avg rating*

```python
popularity_df=popularity_df[popularity_df['rating_num']>=250].sort_values('avg_rating',ascending=False).head(50)
```

```python
popularity_df=popularity_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','rating_num','avg_rating']]
```

> *here*'s are recommanded top 50 books data :

```python
popularity_df
```

#**Collabrative-Filttering-Based-Sys**

**> filtering based on user**

```python
x = mr_ratings.groupby('User-ID').count()['Book-Rating']>200
iq_user = x[x].index
```

```python
fltr_rating= mr_ratings[mr_ratings['User-ID'].isin(iq_user)]
fltr_rating
```

**> filtering based on books**

```python
y =  fltr_rating.groupby('Book-Title').count()['Book-Rating'] >=50
famous_books = y[y].index
```

```python
final_rating = fltr_rating[fltr_rating['Book-Title'].isin(famous_books)]
final_rating
```

**> Build pivot Table**

```python
pt= final_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
pt.fillna(0,inplace=True)
```

```python
pt
```

**> finding nuclic distance of every book which is currently represented as a vector in 810- dimensions to every other books**

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score.shape
```

**> Now we have to build a recommend system that will take book name and return 5 recommanded books**

```python
def recommend(book_name):
  #index fetch
  index = np.where(pt.index==book_name)[0][0]
  similar_items = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]

  for i in similar_items:
    print(pt.index[i[0]])
```

```python
recommend('1984')
```
